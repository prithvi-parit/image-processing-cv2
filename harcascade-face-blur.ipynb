```python
model=cv2.CascadeClassifier('haarcascade_frontalcatface.xml')
```

```python
ret,photo=cap.read()
```

```python
fdetect=model.detectMultiScale(photo)
```

```python
fdetect
```

```python
x1=fdetect[0][0]
```

```python
y1=fdetect[0][1]
```

```python
x2=fdetect[0][2]+x1
```

```python
y2=fdetect[0][3]+y1
```

```python
a=cv2.rectangle(photo,(x1,y1),(x2,y2),[0,255,0],2)
```

```python
cv2.imshow("hi",a)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
while True:
    ret,photo=cap.read()
    fdetect=model.detectMultiScale(photo)
    if len(fdetect)==0:
        print("no face")
    else:
        x1=fdetect[0][0]
        y1=fdetect[0][1]
        x2=fdetect[0][2]+x1
        y2=fdetect[0][3]+y1
        #a=cv2.rectangle(photo,(x1,y1),(x2,y2),[0,255,0],2)
        cphoto=photo[y1:y2,x1:x2]
        bphoto=cv2.blur(cphoto,(50,50))
        photo[y1:y2,x1:x2]=bphoto
        cv2.imshow("hi",photo)
        if cv2.waitKey(10)==13:
            break
cv2.destroyAllWindows()      
    
```
