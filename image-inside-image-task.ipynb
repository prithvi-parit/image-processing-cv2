```python
import cv2
```

```python
cap=cv2.VideoCapture(0)
```

```python
while True:
    ret,photo=cap.read()
    #nphoto=photo[100:400,150:450]
    #nphoto=cv2.cvtColor(photo,cv2.COLOR_BGR2GRAY)
    #nphoto=cv2.cvtColor(photo,cv2.COLOR_BGR2HLS)
    #nphoto=cv2.cvtColor(photo,cv2.COLOR_BGR2HSV)
    #nphoto=cv2.cvtColor(photo,cv2.COLOR_BGR2Lab)
    #nphoto=cv2.cvtColor(photo,cv2.COLOR_BGR2YUV)
    #nphoto=cv2.cvtColor(photo,cv2.COLOR_BGR2YCrCb)
    #nphoto=cv2.cvtColor(photo,cv2.COLOR_BGR2YCR_CB)
    scale_percent = 30 # percent of original size
    width = int(photo.shape[1] * scale_percent / 100)
    height = int(photo.shape[0] * scale_percent / 100)
    dim = (width, height)
   # print(width,height)
    
    resized = cv2.resize(photo, dim, interpolation = cv2.INTER_AREA)
  #  print(resized.shape)
  #  print(photo.shape)
    photo[0:144,0:192]=resized
   # print(nphoto.shape)
    cv2.imshow("my title",photo)
    if cv2.waitKey(100)==13:
        break
cv2.destroyAllWindows()
```

```python
cap.release()
```
